```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import os
```

```python
df = pd.read_csv('./mgdata.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.cluster import KMeans
```

```python
k = 200
model = KMeans(n_clusters = k, random_state = 1234)
model.fit(df[['X','Y']])
df['cluster'] = model.fit_predict(df[['X','Y']])
```

```python
df.head()
```

```python
pd.DataFrame(df['cluster'].value_counts()).head()
```

```python
cluster_centers = model.cluster_centers_
```

```python
df_center = pd.DataFrame(cluster_centers)
df_center.head()
```

```python
import folium
```

```python
map = folium.Map(location=[df_center.loc[:,1].mean(),df_center.loc[:,0].mean()], zoom_start=11)

for n in df_center.index:
    location = [df_center.loc[n, 1], df_center.loc[n, 0]]
    folium.Marker(
        location = location,
    ).add_to(map)
    
for n in df_center.index:
    folium.Circle(
        location=[df_center.loc[n,1], df_center.loc[n,0]],
        #icon=folium.Icon(color='red',icon='ok')
        radius = 100
    ).add_to(map)

map
```
