```python
df = pd.read_csv('./alldata.csv')
```

```python
df2 = df.copy()
```

## 1. 쓰레기 발생량

```python
# 구간 10개로 나누기
df['쓰레기발생량_그룹'] = pd.cut(df['쓰레기발생량'], bins = 10, labels = list(range(1,11,1)))
df
```

## 2. 주택수

```python
# 구간 10개로 나누기
df2['주택수_그룹'] = pd.cut(df2['주택수'], bins = 10, labels = list(range(1,11,1)))
df2
```

## 3. 1인가구 평균 배달일수

```python
# 구간 10개로 나누기
df2['1인가구 평균 배달일수_그룹'] = pd.cut(df2['1인가구 평균 배달일수'], bins = 10, labels = list(range(1,11,1)))
df2
```

## 4. 1인가구 비율

```python
# 구간 10개로 나누기
df2['1인가구 비율_그룹'] = pd.cut(df2['1인가구 비율'], bins = 10, labels = list(range(1,11,1)))
df2
```

## 합계

```python
df2['쓰레기발생량_그룹'] = df2['쓰레기발생량_그룹'].astype('int')
df2['주택수_그룹'] = df2['주택수_그룹'].astype('int')
df2['1인가구 평균 배달일수_그룹'] = df2['1인가구 평균 배달일수_그룹'].astype('int')
df2['1인가구 비율_그룹'] = df2['1인가구 비율_그룹'].astype('int')

df2['sum'] = df2['쓰레기발생량_그룹'] + df2['주택수_그룹'] + df2['1인가구 평균 배달일수_그룹'] + df2['1인가구 비율_그룹']
df2
```

```python
df3 = df2.sort_values('sum', ascending = False)[['지역', 'sum']]
df3
```

```python
df3.to_csv('자치구별 등급.csv', index = False, encoding='utf-8-sig')
```
